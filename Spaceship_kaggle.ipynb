```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"C:\Users\patha\Downloads\spaceship-titanic\train.csv")
df=df.drop(['Name'],axis=1)
df=df.drop(['PassengerId'],axis=1)
df
```

```python
df.isnull().sum()
```

```python
df['HomePlanet']=df['HomePlanet'].fillna(df['HomePlanet'].mode()[0])
df['CryoSleep']=df['CryoSleep'].fillna(df['CryoSleep'].mode()[0])
df['Destination']=df['Destination'].fillna(df['Destination'].mode()[0])
df['Age']=df['Age'].fillna(df['Age'].mean())
df['VIP']=df['VIP'].fillna(df['VIP'].mode()[0])
df['RoomService']=df['RoomService'].fillna(df['RoomService'].median())
df['FoodCourt']=df['FoodCourt'].fillna(df['FoodCourt'].median())
df['ShoppingMall']=df['ShoppingMall'].fillna(df['ShoppingMall'].median())
df['Spa']=df['Spa'].fillna(df['Spa'].median())
df['VRDeck']=df['VRDeck'].fillna(df['VRDeck'].median())
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
label1=le.fit_transform(df['Transported'])
df['Transported']=label1
label2=le.fit_transform(df['HomePlanet'])
df['HomePlanet']=label2
label3=le.fit_transform(df['CryoSleep'])
df['CryoSleep']=label3
label4=le.fit_transform(df['Cabin'])
df['Cabin']=label4
label5=le.fit_transform(df['Destination'])
df['Destination']=label5
label6=le.fit_transform(df['VIP'])
df['VIP']=label6
df
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
X=df.drop(['Transported'],axis=1)
y=df['Transported']
rf=RandomForestClassifier()
rf.fit(X,y)
```

```python
df_test=pd.read_csv(r"C:\Users\patha\Downloads\spaceship-titanic\test.csv")
df_test=df_test.drop(['Name'],axis=1)
df_test=df_test.drop(['PassengerId'],axis=1)
df_test
```

```python
df_test.isnull().sum()
```

```python
df_test['HomePlanet']=df_test['HomePlanet'].fillna(df_test['HomePlanet'].mode()[0])
df_test['CryoSleep']=df_test['CryoSleep'].fillna(df_test['CryoSleep'].mode()[0])
df_test['Destination']=df_test['Destination'].fillna(df_test['Destination'].mode()[0])
df_test['Age']=df_test['Age'].fillna(df_test['Age'].mean())
df_test['VIP']=df_test['VIP'].fillna(df_test['VIP'].mode()[0])
df_test['RoomService']=df_test['RoomService'].fillna(df_test['RoomService'].median())
df_test['FoodCourt']=df_test['FoodCourt'].fillna(df_test['FoodCourt'].median())
df_test['ShoppingMall']=df_test['ShoppingMall'].fillna(df_test['ShoppingMall'].median())
df_test['Spa']=df_test['Spa'].fillna(df_test['Spa'].median())
df_test['VRDeck']=df_test['VRDeck'].fillna(df_test['VRDeck'].median())
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()

label2=le.fit_transform(df_test['HomePlanet'])
df_test['HomePlanet']=label2
label3=le.fit_transform(df_test['CryoSleep'])
df_test['CryoSleep']=label3
label4=le.fit_transform(df_test['Cabin'])
df_test['Cabin']=label4
label5=le.fit_transform(df_test['Destination'])
df_test['Destination']=label5
label6=le.fit_transform(df_test['VIP'])
df_test['VIP']=label6
df_test
```

```python
import numpy as np
y_pred=rf.predict(df_test)
l=[]
y_pred=np.array(y_pred)
for i in y_pred:
    if i==1:
        l.append("FALSE")
    else:
        l.append("TRUE")
y_pred=pd.DataFrame(l)
```

```python
df_test=df_test.drop(['HomePlanet','CryoSleep','Cabin','Destination','Age','VIP','RoomService','FoodCourt','ShoppingMall','Spa','VRDeck'],axis=1)
df_test=pd.concat([df_test, y_pred], axis=1).reindex(df_test.index)
df_test
```
